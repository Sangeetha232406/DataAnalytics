```python
column_names = {
    'Rape': 'Rape Cases',
    'K&A': 'Kidnap and Assault',
    'DD': 'Dowry Deaths',
    'AoW': 'Assault on Women',
    'AoM': 'Assault on Minors',
    'DV': 'Domestic Violence',
    'WT': 'Women Trafficking'
}

df.rename(columns=column_names, inplace=True)
```

```python
pie_plot = ['Rape Cases', 'Kidnap and Assault', 'Dowry Deaths', 'Assault on Women', 'Assault on Minors', 'Domestic Violence', 'Women Trafficking']  # Corrected column name: 'Women Trafficking'
crime_totals = df[pie_plot].sum()
plt.figure(figsize=(10,5))
plt.pie(crime_totals, labels=pie_plot, autopct='%1.1f%%')  # labels and percentage
plt.title('Distribution of Crimes')
plt.show()
```

```python

crime_trend = df.drop(columns=['TotalCrimes'], errors='ignore').groupby('Year').sum()  # Drop 'TotalCrimes'

plt.figure(figsize=(12, 6))
sns.lineplot(data=crime_trend)
plt.title('Trend of Crimes Against Women (2001-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python

filtered_df = df[df['State'] != '#REF!']
state_counts = filtered_df['State'].value_counts().sort_index()

palette = sns.color_palette("icefire", n_colors=len(state_counts))
labels = filtered_df['State'].unique()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
ax1.pie(state_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=palette)
ax1.set_title('State Class Distribution (Pie Chart)')
sns.countplot(x='State', data=filtered_df, ax=ax2, palette=palette)
ax2.set_title('State Class Distribution (Count Plot)')
ax2.set_xlabel('State Class')
ax2.set_ylabel('Count')
ax2.set_xticklabels(labels, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
crime_types =['Rape Cases', 'Kidnap and Assault', 'Dowry Deaths', 'Assault on Women', 'Assault on Minors', 'Domestic Violence', 'Women Trafficking']  # Corrected column name: 'Women Trafficking'
highest_crime_per_state = {}
new_df = df[df['State'] != '#REF!']
state_crime_totals = new_df.groupby('State')[crime_types].sum()
for state in state_crime_totals.index:
    crime_totals = state_crime_totals.loc[state]
    highest_crime_type = crime_totals.idxmax()
    highest_crime_count = crime_totals.max()
    highest_crime_per_state[state] = (highest_crime_type, highest_crime_count)
highest_crime_df = pd.DataFrame.from_dict(highest_crime_per_state, orient='index', columns=['Highest Crime', 'Count'])
print(highest_crime_df)
```

```python
state_crime_totals = new_df.groupby('State')['TotalCrimes'].sum().reset_index()
mean_crime = state_crime_totals['TotalCrimes'].mean()
std_crime = state_crime_totals['TotalCrimes'].std()
threshold = mean_crime + std_crime

# Categorize states based on the threshold
state_crime_totals['CrimeZone'] = state_crime_totals['TotalCrimes'].apply(lambda x: 'High' if x > threshold else 'Low')

# Print the categorized states
print(state_crime_totals[['State', 'CrimeZone']])
plt.figure(figsize=(12, 6))
plt.bar(state_crime_totals['State'], state_crime_totals['TotalCrimes'], color=state_crime_totals['CrimeZone'].map({'High': 'red', 'Low': 'blue'}))
plt.title('Crime Zones in India')
plt.xlabel('State')
plt.ylabel('Total Crimes')
plt.xticks(rotation=90)
plt.show()
```

```python
state_crime_totals = new_df.groupby('State')['TotalCrimes'].sum().reset_index()
top_10_states = state_crime_totals.sort_values('TotalCrimes', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_states['State'], top_10_states['TotalCrimes'])
plt.title('Top 10 States with Highest Crime Rates')
plt.xlabel('State')
plt.ylabel('Total Crimes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
least_crime_states = state_crime_totals.sort_values('TotalCrimes', ascending=True).head(10)
plt.figure(figsize=(12, 6))
plt.plot(least_crime_states['State'], least_crime_states['TotalCrimes'])
plt.title('Top 10 States with Lowest Crime Rates')
plt.xlabel('State')
plt.ylabel('Total Crimes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
state_crime_totals['CrimePercentile'] = state_crime_totals['TotalCrimes'].rank(pct=True)

high_threshold = 0.75
low_threshold = 0.25
state_crime_totals['CrimeZone'] = pd.cut(
    state_crime_totals['CrimePercentile'],
    bins=[0, low_threshold, high_threshold, 1],
    labels=['Low', 'Medium', 'High'],
)
medium_crime_states = state_crime_totals[state_crime_totals['CrimeZone'] == 'Medium']

plt.figure(figsize=(12, 6))
plt.scatter(medium_crime_states['State'], medium_crime_states['TotalCrimes'], s=medium_crime_states['TotalCrimes'] / 1000, alpha=0.5)  # Adjust marker size scaling as needed
plt.title('Medium Range Crime States in India')
plt.xlabel('State')
plt.ylabel('Total Crimes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

data_cleaned = df.drop(columns=["TotalCrimes"], errors="ignore")
data_cleaned.replace("#REF!", pd.NA, inplace=True)
data_cleaned.dropna(inplace=True)

crime_columns = data_cleaned.columns[2:]
data_cleaned[crime_columns] = data_cleaned[crime_columns].apply(pd.to_numeric, errors="coerce")

state_crime_data = data_cleaned.groupby("State")[crime_columns].sum()
state_crime_data.plot(kind="bar", stacked=True, figsize=(15, 8), colormap="tab20")
plt.title("Crimes Against Women by State")
plt.xlabel("State")
plt.ylabel("Number of Crimes")
plt.legend(title="Type of Crime", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
